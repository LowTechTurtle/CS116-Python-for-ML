# Load data:

```python
import gc
import itertools
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(style="whitegrid")

# Load dữ liệu
sales = pd.read_csv('competitive-data-science-predict-future-sales/sales_train.csv')
items = pd.read_csv('competitive-data-science-predict-future-sales/items.csv')
item_categories = pd.read_csv('competitive-data-science-predict-future-sales/item_categories.csv')
shops = pd.read_csv('competitive-data-science-predict-future-sales/shops.csv')
test = pd.read_csv('competitive-data-science-predict-future-sales/test.csv')

sales.head()
```

# Univariate EDA:

```python
# Convert to datetime
sales['date'] = pd.to_datetime(sales['date'], format='%d.%m.%Y')

# Monthly total item sales
monthly_sales = sales.groupby('date_block_num')['item_cnt_day'].sum()

plt.figure(figsize=(12,6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Total item sales per month")
plt.xlabel("Month Number")
plt.ylabel("Total Items Sold")
```

# Multivariate EDA:

```python
# Prepare monthly aggregated data
train_monthly = sales.groupby(['date_block_num', 'shop_id', 'item_id']).agg({
    'item_cnt_day': 'sum'
}).reset_index().rename(columns={'item_cnt_day': 'item_cnt'})

train_monthly = train_monthly.merge(items[['item_id', 'item_category_id']], on='item_id', how='left')

# Add month feature (0 to 11)
train_monthly['month'] = train_monthly['date_block_num'] % 12

# Monthly mean and sum of item sales
gp_month_mean = train_monthly.groupby(['month'], as_index=False)['item_cnt'].mean()
gp_month_sum = train_monthly.groupby(['month'], as_index=False)['item_cnt'].sum()

# Per-category analysis
gp_category_mean = train_monthly.groupby(['item_category_id'], as_index=False)['item_cnt'].mean()
gp_category_sum = train_monthly.groupby(['item_category_id'], as_index=False)['item_cnt'].sum()

# Per-shop analysis
gp_shop_mean = train_monthly.groupby(['shop_id'], as_index=False)['item_cnt'].mean()
gp_shop_sum = train_monthly.groupby(['shop_id'], as_index=False)['item_cnt'].sum()
```

## Xử lý dữ liệu

thử biến đổi đặc trưng bằng nhiều cách (log transform, min-max, z-score) để khi cần dùng cái nào thì dùng, cũng như dùng label encoder để encoding dữ liệu dạng danh mục (ko phân biệt thứ tự)

```python
# 1. Logarithmic transformation for price and count (handle skewness)
sales['log_item_price'] = np.log(sales['item_price'])
sales['log_item_cnt_day'] = np.log(sales['item_cnt_day'] + 1)  # Add 1 to handle zeros

# 2. Min-Max scaling - Normalize data to [0, 1] range
from sklearn.preprocessing import MinMaxScaler

scaler_minmax = MinMaxScaler()
sales['item_price_scaled'] = scaler_minmax.fit_transform(sales[['item_price']])
sales['item_cnt_day_scaled'] = scaler_minmax.fit_transform(sales[['item_cnt_day']])

# 3. Z-score standardization - Mean=0, STD=1
from sklearn.preprocessing import StandardScaler

scaler_standard = StandardScaler()
sales['item_price_standard'] = scaler_standard.fit_transform(sales[['item_price']])
sales['item_cnt_day_standard'] = scaler_standard.fit_transform(sales[['item_cnt_day']])


# Không cần encoding cho shop_id và item_id vì đã là số

# 4. Encoding các danh mục
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
sales['shop_id_encoded'] = label_encoder.fit_transform(sales['shop_id'])
sales['item_id_encoded'] = label_encoder.fit_transform(sales['item_id'])


print(sales.head())
```

# Tổng hợp dữ liệu

chuyển đổi dữ liệu bán hàng hàng ngày thành dữ liệu bán hàng hàng tháng để phù hợp với mục tiêu dự đoán của bài toán

```python
# Tạo ma trận huấn luyện giống như cấu trúc test
def create_testlike_train(sales_train, test=None):
    indexlist = []
    for i in sales_train.date_block_num.unique():
        x = itertools.product(
            [i],
            sales_train.loc[sales_train.date_block_num == i].shop_id.unique(),
            sales_train.loc[sales_train.date_block_num == i].item_id.unique(),
        )
        indexlist.append(np.array(list(x)))
    
    df = pd.DataFrame(
        data=np.concatenate(indexlist, axis=0),
        columns=["date_block_num", "shop_id", "item_id"]
    )

    # Tổng hợp theo tháng
    sales_grouped = sales_train.groupby(["date_block_num", "shop_id", "item_id"]).agg(
        item_cnt_month=pd.NamedAgg(column="item_cnt_day", aggfunc="sum"),
        item_revenue_month=pd.NamedAgg(column="item_revenue_day", aggfunc="sum"),
    ).reset_index()

    df = df.merge(sales_grouped, how="left", on=["date_block_num", "shop_id", "item_id"])

    if test is not None:
        test['date_block_num'] = 34
        test = test.drop(columns="ID")
        df = pd.concat([df, test[["date_block_num", "shop_id", "item_id"]]])

    df['item_cnt_month'] = df['item_cnt_month'].fillna(0)
    df['item_revenue_month'] = df['item_revenue_month'].fillna(0)
    return df

matrix = create_testlike_train(sales, test)
print (matrix.head())
```
